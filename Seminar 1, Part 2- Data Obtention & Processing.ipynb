### Data Visualization & more processing

As you might be related to python ways of visualizing & processing data, you can see how you can quickly and efficiently get company data i the time steps & scales you wish to work in

```python
from yahooquery import Ticker  # Import the Ticker class

aapl = Ticker('aapl')
```

```python
aapl.history()
```

If no arguments are provided, as above, default values will be supplied for both period and interval, which are ytd and 1d, respectively. Additional arguments you can provide to the method are start and end. Start and end dates can be either strings with a date format of yyyy-mm-dd or as a datetime.datetime object.

```python
aapl.history(period='max')
```

```python
aapl.history(start='2019-05-01')  # Default end date is now

# Period options = 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y, ytd, max
# Interval options = 1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk, 1mo, 3mo
```

```python
df=aapl.history(start='2009-05-01', end='2018-12-31') 
df
```

```python
# Save data in memory to disk

df.to_csv('Data_Stock_Date.csv')
df.to_csv('Data_Stock_No_Date.csv', index=False)
```

### Furthering the financial metrics methods

The following methods take a frequency argument.  If nothing is provided, annual data will be returned.  To return quarterly data, pass "q" as an argument. These methods are also good to follow the financials of the company chosen in a summarized scale 

```python
aapl.balance_sheet()  # Defaults to Annual
```

```python
aapl.balance_sheet(frequency="q")
```
